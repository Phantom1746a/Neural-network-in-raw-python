```python
#Base class
class Layer:
  def __init__(self):
    self.input=None
    self.output=None
   # computes the output Y of a layer for a given input X
  def forward(self,input):
    pass
    # computes dE/dX for a given dE/dY (and update parameters if any)
  def backward(self,output_gadient, learning_rate):
    pass


```

```python
#Dense layer class
import numpy as np

class Dense(Layer):
  def __init__(self, input_size, output_size):
    self.weights = np.random.randn(output_size,input_size)
    self.bias = np.random.randn(output_size,1)

 # computes the output Y of a layer for a given input X
  def forward(self, input):
    self.input= input
    return np.dot(self.weights,self.input) + self.bias

# computes dE/dX for a given dE/dY (and update parameters if any)
  def backward(self,output_gadient, learning_rate):
      weight_gradient = np.dot(output_gadient, self.input.T)
      input_gradient = np.dot(self.weights.T,output_gadient)
      self.weights -= learning_rate * weight_gradient
      self.bias -= learning_rate * output_gadient
      return input_gradient

```

```python
# Activation layer Forward plan

class Activation(Layer):
  def __init__(self, activation, activation_prime):
    self.activation = activation
    self.activation_prime = activation_prime


  def forward(self,input):
    self.input =input
    return self.activation(self.input)

  def backward(self, output_gradient, learning_rate):
    return np.multiply(output_gradient, self.activation_prime(self.input))
```

```python
#hyperbolic Tangent function
class Tanh(Activation):
  def __init__(self):
    tanh = lambda x: np.tanh(x)
    tanh_prime = lambda x: 1-np.tanh(x) **2
    super().__init__(tanh,tanh_prime)
```

```python
#mean squared error
def mse(y_true, y_pred):
  return np.mean(np.power(y_true-y_pred,2))


def mse_prime(y_true,y_pred):
  return 2 * (y_pred-y_true)/np.size(y_true)
```

```python
#Implementing Neural Network
# XOR - Exclusive OR
X= np.reshape([[0,0],[0,1],[1,0],[1,1]],(4,2,1))
Y= np.reshape([[0],[0],[0],[1]],(4,1,1))
network =[Dense(2,3),
          Tanh(),
          Dense(3,1),
          Tanh()
          ]
epochs =30000
learning_rate= 0.1

#train
for e in range(epochs):
  error =0
  for x,y in zip(X,Y):
    #forward
    output=x
    for layer in network:
      output = layer.forward(output)

      #error
      error += mse(y,output)
      grad = mse_prime(y, output)
    for layer in reversed(network):
        grad = layer.backward(grad, learning_rate)


  error /= len(X)
  print('%d/%d, error=%f' % (e+1, epochs , error))


```

```python
# Forward pass for the trained neural network
def forward_pass(x, network):
    output = x
    for layer in network:
        output = layer.forward(output)
    return output

# Test the trained network on XOR inputs
for x, y_target in zip(X, Y):
    # Perform a forward pass through the trained network
    y_pred = forward_pass(x, network)

    print(f"Input: {x.flatten()}, Target: {y_target.flatten()}, Predicted: {y_pred.flatten()}")
```
